```python
import matplotlib.pyplot as plt
import scipy as sp
import numpy as np
import pandas as pd

%pylab
%load_ext autoreload
%autoreload
%matplotlib inline
```

```python
n = 20 # number of vectors to split up on circle
radius = 5 # radius of circle to generate


x_0 = radius
y_0 = 0
v_circle = [] # 
v_star = [] # star arrangement of vectors
for i in xrange(1,n+1):
    angle = i*(2*sp.pi/n)
    x = radius*sp.cos(angle)
    y = radius*sp.sin(angle)

    v_circle.append([x_0,y_0,x-x_0, y-y_0])
    v_star.append([0,0, x, y])
    
    x_0 = x
    y_0 = y
    
    
x, y, u, v = zip(*v_circle)
xs, ys, us, vs = zip(*v_star)

fig, axes = plt.subplots(1,2,figsize=(20, 10))

axes[0].quiver(x,y,u,v,angles='xy',scale_units='xy',scale=1)
axes[0].set_xlim([-10,10])
axes[0].set_ylim([-10,10])
axes[0].set_title("Circle vectors")

axes[1].quiver(xs,ys,us,vs,angles='xy',scale_units='xy',scale=1)
axes[1].set_xlim([-10,10])
axes[1].set_ylim([-10,10])
axes[1].set_title("Star vectors")
plt.draw()
plt.show()
```
